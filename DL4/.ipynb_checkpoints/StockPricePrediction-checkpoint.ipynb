```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

# 3. Data Collection

### 3.0 import the data

```python
dataset_train = pd.read_csv('Google_Stock_Price_Train.csv')
```

```python
dataset_train.head()
```

```python
#keras only takes numpy array
training_set = dataset_train.iloc[:, 1: 2].values
```

```python
training_set.shape
```

### 3.1 Feature scaling

```python
sc = MinMaxScaler(feature_range = (0, 1))
#fit: get min/max of train data
training_set_scaled = sc.fit_transform(training_set)
```

### 3.2 Data structure creation

- taking the reference of past 60 days of data to predict the future stock price.  
- It is observed that taking 60 days of past data gives us best results.  
- In this data set 60 days of data means 3 months of data.  
- Every month as 20 days of Stock price.   
- X train will have data of 60 days prior to our date and y train will have data of one day after our date

```python
## 60 timesteps and 1 output
X_train = []
y_train = []
for i in range(60, len(training_set_scaled)):
    X_train.append(training_set_scaled[i-60: i, 0])
    y_train.append(training_set_scaled[i, 0])

X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
X_train.shape
```

```python
y_train.shape
```

### 3.3 Data reshaping

```python
X_train = np.reshape(X_train, newshape = 
                     (X_train.shape[0], X_train.shape[1], 1))
```

1. Number of stock prices - 1449  
2. Number of time steps - 60  
3. Number of Indicator - 1  

```python
X_train.shape
```

# 4. Exploratory Data Analysis

```python
plt.figure(figsize=(18, 8))
plt.plot(dataset_train['Open'])
plt.title("Google Stock Open Prices")
plt.xlabel("Time (oldest -> latest)")
plt.ylabel("Stock Open Price")
plt.show()
```

```python
plt.figure(figsize=(18, 8))
plt.plot(dataset_train['Low'])
plt.title("Google Stock Low Prices")
plt.xlabel("Time (oldest -> latest)")
plt.ylabel("Stock Lowest Price")
plt.show()
```

# 5. Create & Fit Model

### 5.1 Create model

```python
regressor = Sequential()
#add 1st lstm layer
regressor.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 1)))
regressor.add(Dropout(rate = 0.2))

##add 2nd lstm layer: 50 neurons
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(rate = 0.2))

##add 3rd lstm layer
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(rate = 0.2))

##add 4th lstm layer
regressor.add(LSTM(units = 50, return_sequences = False))
regressor.add(Dropout(rate = 0.2))

##add output layer
regressor.add(Dense(units = 1))
```

```python
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

### 5.2 Model fit

```python
regressor.fit(x = X_train, y = y_train, batch_size = 32, epochs = 100)
```

```python
#vertical concat use 0, horizontal uses 1
dataset_total = pd.concat((dataset_train['Open'], dataset_test['Open']), 
                          axis = 0)
##use .values to make numpy array
inputs = dataset_total[len(dataset_total) - len(dataset_test) - 60:].values
```

### 6.3 Reshape and scale

```python
#reshape data to only have 1 col
inputs = inputs.reshape(-1, 1)

#scale input
inputs = sc.transform(inputs)
```

### 6.4 Create test data strucutre

```python
X_test = []
for i in range(60, len(inputs)):
    X_test.append(inputs[i-60:i, 0])
X_test = np.array(X_test)
#add dimension of indicator
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

# 7. Model prediction

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout


dataset_train = pd.read_csv('Google_Stock_Price_Train.csv')

dataset_train.head()


#keras only takes numpy array
training_set = dataset_train.iloc[:, 1: 2].values

training_set.shape


sc = MinMaxScaler(feature_range = (0, 1))
#fit: get min/max of train data
training_set_scaled = sc.fit_transform(training_set)



## 60 timesteps and 1 output
X_train = []
y_train = []
for i in range(60, len(training_set_scaled)):
    X_train.append(training_set_scaled[i-60: i, 0])
    y_train.append(training_set_scaled[i, 0])

X_train, y_train = np.array(X_train), np.array(y_train)


X_train.shape



y_train.shape


X_train = np.reshape(X_train, newshape = 
                     (X_train.shape[0], X_train.shape[1], 1))




X_train.shape



plt.figure(figsize=(18, 8))
plt.plot(dataset_train['Open'])
plt.title("Google Stock Open Prices")
plt.xlabel("Time (oldest -> latest)")
plt.ylabel("Stock Open Price")
plt.show()



plt.figure(figsize=(18, 8))
plt.plot(dataset_train['Low'])
plt.title("Google Stock Low Prices")
plt.xlabel("Time (oldest -> latest)")
plt.ylabel("Stock Lowest Price")
plt.show()




regressor = Sequential()
#add 1st lstm layer
regressor.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 1)))
regressor.add(Dropout(rate = 0.2))

##add 2nd lstm layer: 50 neurons
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(rate = 0.2))

##add 3rd lstm layer
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(rate = 0.2))

##add 4th lstm layer
regressor.add(LSTM(units = 50, return_sequences = False))
regressor.add(Dropout(rate = 0.2))

##add output layer
regressor.add(Dense(units = 1))





regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')




regressor.fit(x = X_train, y = y_train, batch_size = 32, epochs = 100)


dataset_test = pd.read_csv('Google_Stock_Price_Test.csv')


dataset_test.head()




#keras only takes numpy array
real_stock_price = dataset_test.iloc[:, 1: 2].values
real_stock_price.shape




#vertical concat use 0, horizontal uses 1
dataset_total = pd.concat((dataset_train['Open'], dataset_test['Open']), 
                          axis = 0)
##use .values to make numpy array
inputs = dataset_total[len(dataset_total) - len(dataset_test) - 60:].values



#reshape data to only have 1 col
inputs = inputs.reshape(-1, 1)

#scale input
inputs = sc.transform(inputs)


len(inputs)





X_test = []
for i in range(60, len(inputs)):
    X_test.append(inputs[i-60:i, 0])
X_test = np.array(X_test)
#add dimension of indicator
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))



X_test.shape



predicted_stock_price = regressor.predict(X_test)


#inverse the scaled value
predicted_stock_price = sc.inverse_transform(predicted_stock_price)



##visualize the prediction and real price
plt.plot(real_stock_price, color = 'red', label = 'Real price')
plt.plot(predicted_stock_price, color = 'blue', label = 'Predicted price')

plt.title('Google price prediction')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()


```

This code implements a time series forecasting model using LSTM (Long Short-Term Memory) neural network architecture to predict the stock prices of Google. Here's a theoretical explanation of each part:

1. **Data Preprocessing**:
   - The code first imports the necessary libraries and loads the dataset containing historical stock prices.
   - It scales the training data using Min-Max scaling to normalize the values between 0 and 1, which helps improve the performance of the neural network.
   - The training data is then split into input sequences (X_train) and corresponding output labels (y_train) with a specified window size (60 in this case).

2. **Model Architecture**:
   - The model architecture consists of a stack of LSTM layers followed by dropout layers to prevent overfitting.
   - Four LSTM layers with 50 neurons each are stacked together. The `return_sequences` parameter is set to `True` for all layers except the last one to return sequences rather than single outputs.
   - After the LSTM layers, a dense output layer with a single neuron is added to predict the next stock price.

3. **Model Compilation and Training**:
   - The model is compiled using the Adam optimizer and mean squared error (MSE) loss function.
   - It is then trained on the preprocessed training data (X_train and y_train) for 100 epochs with a batch size of 32.

4. **Testing and Prediction**:
   - The test dataset containing the real stock prices is loaded and preprocessed similarly to the training data.
   - Input sequences for the test data (X_test) are created using the same window size (60).
   - The trained model is then used to predict the stock prices for the test data.
   - The predicted stock prices are inverse-transformed to obtain the actual price values.

5. **Visualization**:
   - The real stock prices and the predicted prices are plotted on a graph to visualize the model's performance.

Long Short-Term Memory (LSTM) is a type of recurrent neural network (RNN) architecture that is well-suited for processing and making predictions on sequences of data. It addresses some of the limitations of traditional RNNs, such as the vanishing gradient problem, which occurs when gradients become extremely small during training, making it difficult for the model to learn long-term dependencies in the data.

LSTM networks are designed to maintain information over long periods of time, allowing them to effectively capture dependencies in sequential data. They achieve this through the use of a memory cell and several gating mechanisms that control the flow of information. The key components of an LSTM unit include:

1. **Memory Cell**: The memory cell is responsible for storing information over time. It can be thought of as the "memory" of the LSTM unit.

2. **Forget Gate**: This gate determines which information in the memory cell should be discarded or forgotten. It takes as input the current input and the previous hidden state, and outputs a number between 0 and 1 for each element in the memory cell. A value of 1 means "keep this information" and a value of 0 means "forget this information".

3. **Input Gate**: This gate determines which new information should be added to the memory cell. It takes as input the current input and the previous hidden state, and outputs a number between 0 and 1 for each element in the memory cell, representing the importance of each new piece of information.

4. **Cell State Update**: This step updates the memory cell by combining the information from the forget gate (forgetting old information) and the input gate (adding new information).

5. **Output Gate**: This gate determines which information from the memory cell should be passed on to the next time step. It takes as input the current input and the previous hidden state, and outputs a number between 0 and 1 for each element in the memory cell, representing the relevance of each piece of information for the current prediction.

LSTM networks are particularly effective for tasks involving time series data, natural language processing, speech recognition, and other sequential data problems where long-term dependencies are important.
